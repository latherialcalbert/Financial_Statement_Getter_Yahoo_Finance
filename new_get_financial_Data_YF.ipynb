```python
import pandas as pd
import requests

# URL of the Wikipedia page containing the S&P 500 tickers
url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"

# Send a request to the URL and get the response
response = requests.get(url)

# Use pandas to read the tables from the HTML content of the page
tables = pd.read_html(response.text)

# The first table on the page contains the S&P 500 tickers
sp500_table = tables[0]

# Extract the 'Symbol' column from the table
sp500_tickers = sp500_table['Symbol'].tolist()

# Print the list of S&P 500 tickers
print(sp500_tickers)
```

```python
import yfinance as yf
import pandas as pd
import os

# List of S&P 500 tickers

# Create the incomestatements folder if it doesn't exist
os.makedirs('balancesheets', exist_ok=True)

# Set to keep track of saved tickers
saved_tickers = set()

# Loop through the list of tickers
for ticker in sp500_tickers:
    # Skip if the ticker has already been saved
    if ticker in saved_tickers:
        print(f'Income statement data for {ticker} already saved.')
#         continue

    try:
        # Get financial data (income statement) for the current ticker (AAPL)
        financials = yf.Ticker(ticker).balance_sheet

        # Transpose the financial data and print it
        transposed_financials = financials.T
        print(f'Income statement data for {ticker} (Transposed):\n{transposed_financials}')

        # Save the transposed financial data to a CSV file with the ticker symbol in the filename
        csv_filename = f'balancesheets/balance_sheet_{ticker}.csv'
        transposed_financials.to_csv(csv_filename, index=True)  # Set index to True to include row names

        saved_tickers.add(ticker)
        print(f'Transposed income statement data for {ticker} saved to {csv_filename}')
    except Exception as e:
        print(f'Error saving transposed income statement data for {ticker}: {e}')
```
